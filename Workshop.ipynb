## Feature Engineering

### Convertir variable numérica continua a categórica

Creamos una nueva columna para las **bandas de edad**. Esto convierte una variable numérica continua en una variable categórica ordinal.

```python
# Creamos una nueva columna que tenga todas las edades por categoría: 0-10:1, 10-20:2, 20-40:3, 40-60:4, 60-80:5
bins = [0, 10, 20, 40, 60, 80]
group_names = [1, 2, 3, 4, 5]

df['age_categories'] = pd.cut(df['age'], bins, labels=group_names)

df[['age', 'age_categories']].head()
```

```python
df.drop("age", inplace=True, axis=1)

df.head(3)
```

## Train, test

```python
# Asigno las variables train test split

# Variable target solo la columna que voy a predecir
y = df.survived

df.drop(columns="survived", inplace=True)

X_train, X_test, y_train, y_test = train_test_split(df,y, test_size= 0.2)
```

## Modelado y evaluación

 * Los modelos de **regresión** se utilizan cuando la variable objetivo es **cuantitativa**: 
  - salarios
  - emisiones de gases
  - edad de la persona en una foto
  - ...
 * Los modelos de **clasificación** se utilizan cuando la variable objetivo es **cualitativa**: 
  - sobrevivir (o no) al Titanic
  - devolver (o no) un préstamo
  - identificar un perro (o no) en una foto
  - decidir cuál de 3 especies de plantas es ésta
  - ...

**Logistic Regression**

- Mide la relación entre una variable dependiente categórica (a predecir) y una o más variables independientes (predictoras)

- Lo realiza estimando las probabilidades de cada categoría utilizando una función logística, que es la distribución logística acumulativa.



```python
model = LogisticRegression() 
```

```python
model.fit(X_train, y_train)
```

```python
print(model.coef_)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

**Accuracy** 

La exactitud (accuracy) mide el porcentaje de casos que el modelo ha acertado. 

$$accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

¡Cuidado con usar Accuracy con datos no balanceados! Ejemplo: en un conjunto de datos con 990 positivos y 10
negativos, un clasificador que predice siempre “positivo”,
tendría un 0.99 de tasa de aciertos ¡Y sin embargo es malo!



**Precision**

Es la fracción de eventos positivos predichos que son realmente positivos. 

$$precision = \frac{TP}{TP + FP}$$    

Imagina que un sistema antispam tiene una precision baja y te marca un correo como spam aunque éste no lo es, y terminas no leyendo la invitación a la boda de tu hermana.

O en un ejemplo de captación de clientes,  responde a la pregunta ¿qué porcentaje de los clientes que contactemos estarán interesados?

**Recall**

La métrica de exhaustividad nos va a informar sobre la cantidad de positivos que el modelo es capaz de identificar. 

$$recall = \frac{TP}{TP + FN}$$

En un ejemplo de retención de clientes, responde a la pregunta ¿qué porcentaje de los clientes que están interesados somos capaces de identificar?


**F1 Score**

El valor F1 se utiliza para combinar las medidas de precision y recall en un sólo valor. Esto es práctico porque hace más fácil el poder comparar el rendimiento combinado de la precisión y la exhaustividad entre varias soluciones.

F1 se calcula haciendo la media armónica entre la precisión y la exhaustividad:

$$F1 = 2 \cdot \frac{precision \cdot recall}{precision + recall}$$

```python
print("Accuracy: ", accuracy_score(y_test,y_pred))
print("Precision: ", precision_score(y_test,y_pred))
print("Recall: ", recall_score(y_test,y_pred))
print("F1: ", f1_score(y_test, y_pred))
```

Nuestro modelo predice el 84% del tiempo, la supervivencia de los pasajeros correctamente (precision). 

El recall nos dice que predijo que sobreviviría el 77 % de las personas que realmente sobrevivieron.

## Otros modelos de clasificación

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models={
    'dis': LinearDiscriminantAnalysis(),
    'neighbors': KNeighborsClassifier(),
    'naive': GaussianNB(),
    'tree': DecisionTreeClassifier(),
    'svm': SVC(),
    'log': LogisticRegression(max_iter=10000)
}
```

```python
for name, model in models.items():
    print("entrenando ", name)
    model.fit(X_train,y_train)
```

```python
## Hacemos predicciones y medimos

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"--------{name}--------")
    print("Accuracy: ", accuracy_score(y_test,y_pred))
    print("Precision: ", precision_score(y_test,y_pred))
    print("Recall: ", recall_score(y_test,y_pred))
    print("F1: ", f1_score(y_test, y_pred))
```
